```python
data={}
for i in np.unique(stock_5_index['stkcode'].values):
    data[i]=stock_5_index[stock_5_index['stkcode']==i]
    data[i]['stkdret']=np.log(data[i]['stkclose'])-np.log(data[i]['stkclose'].shift(periods=1))
    data[i]['indexdret']=np.log(data[i]['indexclose'])-np.log(data[i]['indexclose'].shift(periods=1))
    data[i].dropna(inplace=True)
    data[i].to_excel(f'{i}.xlsx',index=False)
data
```
