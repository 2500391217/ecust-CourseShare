```python
stock=pd.read_excel('stock.xlsx')
stock
```

```python
data=pd.read_excel('mkt.xlsx')
data
```

```python
data=pd.merge(left=data,right=stock,on='date',how='inner')
data
```

```python
data['date']=pd.to_datetime(data['date'],format='%Y-%m-%d')
data['YYYYmm']=data['date'].dt.strftime('%Y%m')
data
```

```python
data.drop(columns=['date'],inplace=True)
data
```

```python
p1 = data.iloc[:, 2].values
p2 = data.iloc[:, 3].values
p3 = data.iloc[:, 4].values
p4 = data.iloc[:, 5].values
p5 = data.iloc[:, 6].values
p6 = data.iloc[:, 7].values
rf = data.iloc[:, 1].values
mkt = data.iloc[:, 0].values
p1
```

```python
r1 = np.log(p1[1:]) - np.log(p1[:-1])
r2 = np.log(p2[1:]) - np.log(p2[:-1])
r3 = np.log(p3[1:]) - np.log(p3[:-1])
r4 = np.log(p4[1:]) - np.log(p4[:-1])
r5 = np.log(p5[1:]) - np.log(p5[:-1])
r6 = np.log(p6[1:]) - np.log(p6[:-1])
r1
```

```python
rexc1 = r1 - rf[1:]
rexc2 = r2 - rf[1:]
rexc3 = r3 - rf[1:]
rexc4 = r4 - rf[1:]
rexc5 = r5 - rf[1:]
rexc6 = r6 - rf[1:]
rexc1
```

```python
R = np.concatenate([rexc1[:, None],
                    rexc2[:, None],
                    rexc3[:, None],
                    rexc4[:, None],
                    rexc5[:, None],
                    rexc6[:, None]],
                   axis=1)
R
```

```python
Cov_Sample = np.mat(np.cov(R, rowvar=False))
Cov_Sample
```

```python
X = np.mat(np.concatenate([np.ones((len(mkt)-1, 1)), mkt[1:, None]], axis=1))
X
```

```python
Y = np.mat(R)
AB_hat = (X.T*X).I*(X.T*Y)
ALPHA = AB_hat[0]#这里取的是AB_hat的第一行，总之就是常数项的系数
BETA = AB_hat[1]#这里取的是AB_hat的第二行，总之就是市场超额收益率的系数
RESD = Y - X*AB_hat
covfactor = np.cov(mkt[1:])
covresidual = np.diag(np.diag(np.cov(RESD, rowvar=False)))
Cov_Factor = BETA.T*covfactor*BETA + covresidual
Cov_Factor##方法3：因子模型估计法
```

```python
c = 0.5
Cov_Shrink = c*Cov_Sample + (1-c)*Cov_Factor
Cov_Shrink
```

```python
uhat = np.mean(R, axis=0)
A = np.mat(np.concatenate([uhat[:, None], np.ones((len(uhat), 1))], axis=1)).T
up = np.mean(uhat)
b = np.mat(np.array([up, 1])[:, None])
omega_Shrink = Cov_Shrink.I*A.T*(A*Cov_Shrink.I*A.T).I*b
print(omega_Shrink)
omega_Sample = Cov_Sample.I*A.T*(A*Cov_Sample.I*A.T).I*b
print()
print(omega_Sample)
```

```python
df_shrink = pd.DataFrame(columns=data.columns.values[2:8])
df_sample=pd.DataFrame(columns=data.columns.values[2:8])
df_factor=pd.DataFrame(columns=data.columns.values[2:8])
for i in range(67):
    p1 = data.iloc[:i+100, 2].values
    p2 = data.iloc[:i+100, 3].values
    p3 = data.iloc[:i+100, 4].values
    p4 = data.iloc[:i+100, 5].values
    p5 = data.iloc[:i+100, 6].values
    p6 = data.iloc[:i+100, 7].values
    rf = data.iloc[:i+100, 1].values
    mkt = data.iloc[:i+100, 0].values
    r1 = np.log(p1[1:]) - np.log(p1[:-1])
    r2 = np.log(p2[1:]) - np.log(p2[:-1])
    r3 = np.log(p3[1:]) - np.log(p3[:-1])
    r4 = np.log(p4[1:]) - np.log(p4[:-1])
    r5 = np.log(p5[1:]) - np.log(p5[:-1])
    r6 = np.log(p6[1:]) - np.log(p6[:-1])
    rexc1 = r1 - rf[1:i+101]
    rexc2 = r2 - rf[1:i+101]
    rexc3 = r3 - rf[1:i+101]
    rexc4 = r4 - rf[1:i+101]
    rexc5 = r5 - rf[1:i+101]
    rexc6 = r6 - rf[1:i+101]
    R = np.concatenate([rexc1[:, None],
                        rexc2[:, None],
                        rexc3[:, None],
                        rexc4[:, None],
                        rexc5[:, None],
                        rexc6[:, None]],
                       axis=1)
    Cov_Sample = np.mat(np.cov(R, rowvar=False))
    X = np.mat(np.concatenate([np.ones((len(mkt)-1, 1)), mkt[1:, None]], axis=1))
    Y = np.mat(R)
    AB_hat = (X.T*X).I*(X.T*Y)
    ALPHA = AB_hat[0]#这里取的是AB_hat的第一行，总之就是常数项的系数
    BETA = AB_hat[1]#这里取的是AB_hat的第二行，总之就是市场超额收益率的系数
    RESD = Y - X*AB_hat
    covfactor = np.cov(mkt[1:])
    covresidual = np.diag(np.diag(np.cov(RESD, rowvar=False)))
    Cov_Factor = BETA.T*covfactor*BETA + covresidual
    c = 0.5
    Cov_Shrink = c*Cov_Sample + (1-c)*Cov_Factor
    uhat = np.mean(R, axis=0)
    A = np.mat(np.concatenate([uhat[:, None], np.ones((len(uhat), 1))], axis=1)).T
    up = np.mean(uhat)
    b = np.mat(np.array([up, 1])[:, None])
    omega_Shrink = Cov_Shrink.I*A.T*(A*Cov_Shrink.I*A.T).I*b
    omega_Factor = Cov_Factor.I*A.T*(A*Cov_Factor.I*A.T).I*b
    omega_Sample = Cov_Sample.I*A.T*(A*Cov_Sample.I*A.T).I*b
    newrow_shirink=np.asarray(omega_Shrink)
    newrow_shirink=newrow_shirink.reshape(-1)
    newrow_sample=np.asarray(omega_Sample)
    newrow_sample=newrow_sample.reshape(-1)
    newrow_factor=np.asarray(omega_Factor)
    newrow_factor=newrow_factor.reshape(-1)
    df_shrink.loc[len(df_shrink)]=newrow_shirink
    df_sample.loc[len(df_sample)]=newrow_sample
    df_factor.loc[len(df_factor)]=newrow_factor
df_shrink
```
